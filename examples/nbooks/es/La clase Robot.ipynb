# La clase `Robot`

La clase `Robot` es una clase que forma parte del `core` de `moro`, con esta se define un manipulador serial utilizando sus parámetros de Denavit-Hartenberg.

La sintaxis básica de `Robot` es:

```python
r = Robot(DH1, DH2, ..., DHN)
```

Donde `DH1`, `DH2`, ..., `DHN` son tuplas de la forma:

```python
(a_i, alpha_i, d_i, theta_i)
```

Siendo `a_i`, `alpha_i`, `d_i` y `theta_i`, los parámetros de Denavit-Hartenberg.

Antes de comenzar con el ejemplo, vamos a importar la librería:

````python
from moro import *
````

## Ejemplo. Manipulador RR

En la figura se muestra un manipulador RR y su tabla de parámetros DH. 

<img src="img/rr_robot_dh.svg">

Para definirlo hacemos lo siguiente:

````python
RR = Robot((l1,0,0,q1), (l2,0,0,q2))
RR
````

### Calculando la cinemática directa

Para calcular la matriz de la cinemática directa ($ T_n^0 $) podemos acceder al atributo `T`:

````python
RR.T
````

El conjunto de matrices de la forma $ T_i^{i-1} $ se pueden obtener mediante el atributo `Ts`:

````python
RR.Ts
````

### Calculando velocidades 

La matriz jacobiana del extremo del manipulador puede obtenerse con el atributo `J`:

````python
RR.J
````

Naturalmente, para calcular las velocidades lineales y angulares podemos multiplicar la jacobiana por el vector de velocidades articulares, es decir:

````python
RR.J * Matrix([q1p, q2p])
````

Los primeros tres elementos corresponden a la velocidad lineal del extremo ($\boldsymbol{v}_{O_2}$) y los últimos tres a la velocidad angular del último eslabón ($\boldsymbol{\omega}_2$).

La matriz jacobiana para un punto específico puede determinarse utilizando el método `J_point`. A continuación, se muestra cómo calcular la matriz jacobiana del origen del sistema $\{1\}$:

````python
RR.J_point([0,0,0], 1)
````
