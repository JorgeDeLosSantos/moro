```python
def robot_x1():
    R = Robot((0,pi/2,d1,q1), (0,pi/2,0,q2), (0,0,q3,0,"p"))
    R.inertia_tensors = None
    R.masses = None
    R.gravity = (0,0,-g)
    R.cm_positions = [(0,-(d1-lc1),0), 
                        (0,0,lc2), 
                        (0,0,-lc3)]
    X = R._compute_inertia_matrix()
    return X

def robot_x2():
    R = Robot((0,pi/2,d1,q1), (0,pi/2,0,q2), (0,0,q3,0,"p"))
    R.inertia_tensors = None
    R.masses = None
    R.gravity = (0,0,-g)
    R.cm_positions = [(0,-(d1-lc1),0), 
                        (0,0,lc2), 
                        (0,0,-lc3)]
    X = R._compute_inertia_matrix_2()
    return X

def robot_1():
    R = Robot((l1,0,0,q1), (l2,0,0,q2))
    R.inertia_tensors = None
    R.masses = None
    R.gravity = (0,-g,0)
    R.cm_positions = [(-(l1-lc1),0,0),(-(l2-lc2),0,0)]
    X = R._compute_inertia_matrix()
    return X

def robot_2():
    R = Robot((l1,0,0,q1), (l2,0,0,q2))
    R.inertia_tensors = None
    R.masses = None
    R.gravity = (0,-g,0)
    R.cm_positions = [(-(l1-lc1),0,0),(-(l2-lc2),0,0)]
    X = R._compute_inertia_matrix_2()
    return X
```
